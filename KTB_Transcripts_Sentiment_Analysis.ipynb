```python
!pip install PyMuPDF

import fitz  # PyMuPDF
import json
import os

# Function to extract text from a single PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as pdf:
        for page_num in range(pdf.page_count):
            page = pdf[page_num]
            text += page.get_text()
    return text

# Loop through uploaded PDFs and save the extracted text in individual text files
text_data = {}
for pdf_name in uploaded.keys():
    text = extract_text_from_pdf(pdf_name)
    text_data[pdf_name] = text
    # Save each PDF text to a separate text file
    with open(f'{pdf_name}.txt', 'w') as txt_file:
        txt_file.write(text)
    print(f"Text from {pdf_name} saved to {pdf_name}.txt")
```

```python
import os
import json
from datetime import datetime

# Function to process text file into JSON format
def process_text_file_to_json(text_file_path, metadata):
    try:
        # Read text content from the file
        with open(text_file_path, 'r', encoding='utf-8') as text_file:
            text_content = text_file.read()

        # Ensure the file is not empty
        if not text_content.strip():
            print(f"Skipping {text_file_path}: No content found.")
            return None

        # Split the content into sentences
        sentences = text_content.split('. ')  # You can use NLTK for better sentence tokenization

        # Create JSON-like data
        call_data = {
            "Company Name": metadata.get("Company Name"),
            "Date": metadata.get("Date"),
            "Month": metadata.get("Month"),
            "Year": metadata.get("Year"),
            "Ticker Symbol": metadata.get("Ticker Symbol"),
            "Sentences": sentences
        }
        return call_data
    except Exception as e:
        print(f"An error occurred while processing {text_file_path}: {e}")
        return None

# Define the directory where your text files are saved
text_files_dir = "/content"  # Change if your files are saved in a different directory

# Example metadata for 20 files
metadata_list = [
    {
        "Company Name": f"Company {i+1}",
        "Date": str(datetime.now().date()),
        "Month": datetime.now().strftime("%B"),
        "Year": datetime.now().year,
        "Ticker Symbol": f"TICK{i+1}"
    }
    for i in range(20)
]

# List to store JSON data
json_data_list = []

# Get all text files from the directory
text_files = [f for f in os.listdir(text_files_dir) if f.endswith(".txt")]

# Process each text file
for i, text_file in enumerate(text_files):
    if i >= 20:  # Process only the first 20 files
        break

    text_file_path = os.path.join(text_files_dir, text_file)

    # Use metadata for the current file
    metadata = metadata_list[i]

    # Process the text file
    json_data = process_text_file_to_json(text_file_path, metadata)
    if json_data:
        json_data_list.append(json_data)

# Display the processed data in memory
print(f"Processed {len(json_data_list)} text files.")
print(json.dumps(json_data_list[0], indent=4))  # Print the JSON for the first file

```

```python
# Save each JSON object as a separate file
output_dir = "/content/json_files"  # Directory to save JSON files
os.makedirs(output_dir, exist_ok=True)  # Create directory if it doesn't exist

for i, json_data in enumerate(json_data_list):
    output_file_path = os.path.join(output_dir, f"output_{i+1}.json")
    with open(output_file_path, 'w', encoding='utf-8') as json_file:
        json.dump(json_data, json_file, indent=4)
    print(f"Saved JSON file: {output_file_path}")
```

```python
import pandas as pd

# List to store DataFrames
dataframes = []

# Convert each JSON object to a DataFrame
for i, json_data in enumerate(json_data_list):
    # Create a DataFrame for each JSON object
    df = pd.DataFrame(json_data["Sentences"], columns=["Sentence"])
    df["Company Name"] = json_data["Company Name"]
    df["Date"] = json_data["Date"]
    df["Month"] = json_data["Month"]
    df["Year"] = json_data["Year"]
    df["Ticker Symbol"] = json_data["Ticker Symbol"]

    # Append to the list of DataFrames
    dataframes.append(df)
    print(f"DataFrame {i+1} created with {len(df)} rows.")

# Check the first DataFrame
print(dataframes[0].head())
```

```python
!pip install sentence-transformers

from sentence_transformers import SentenceTransformer
import numpy as np

# Load the pre-trained SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Add embeddings to all DataFrames
for i, df in enumerate(dataframes):
    try:
        # Generate sentence embeddings
        df["Embedding"] = df["Sentence"].apply(lambda x: model.encode(x).tolist())
        print(f"Embeddings created for DataFrame {i+1} with {len(df)} sentences.")
    except Exception as e:
        print(f"Error processing DataFrame {i+1}: {e}")

# Display the updated first DataFrame
print(dataframes[0].head())
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Perform PCA to reduce embeddings to 2 dimensions
for i, df in enumerate(dataframes):
    try:
        # Convert the Embedding column (lists) to a numpy array
        embeddings = np.array(df["Embedding"].tolist())

        # Apply PCA
        pca = PCA(n_components=2)
        reduced_embeddings = pca.fit_transform(embeddings)

        # Add the reduced dimensions to the DataFrame
        df["D1"] = reduced_embeddings[:, 0]
        df["D2"] = reduced_embeddings[:, 1]

        print(f"Dimensionality reduction applied to DataFrame {i+1}")
    except Exception as e:
        print(f"Error processing DataFrame {i+1}: {e}")

# Display the updated first DataFrame
print(dataframes[0].head())

```

```python
!pip install transformers

from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
import numpy as np

# Load the pre-trained FinBERT model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")
```

```python
# Define a function to classify sentiment
def classify_sentiment(sentence):
    try:
        # Tokenize the sentence
        inputs = tokenizer(sentence, return_tensors="pt", truncation=True, max_length=512)

        # Perform inference with the model
        outputs = model(**inputs)

        # Get softmax probabilities
        probs = softmax(outputs.logits.detach().numpy()[0])

        # Sentiment labels: [Negative, Neutral, Positive]
        sentiment_labels = ["Negative", "Neutral", "Positive"]

        # Get the sentiment with the highest probability
        sentiment = sentiment_labels[np.argmax(probs)]
        return sentiment
    except Exception as e:
        print(f"Error processing sentence: {e}")
        return "Error"

# Perform sentiment classification for each DataFrame
for i, df in enumerate(dataframes):
    try:
        # Apply sentiment classification to each sentence
        df["Sentiment"] = df["Sentence"].apply(classify_sentiment)
        print(f"Sentiment classification completed for DataFrame {i+1}")
    except Exception as e:
        print(f"Error processing DataFrame {i+1}: {e}")

# Display the updated first DataFrame
print(dataframes[0].head())
```

```python
import matplotlib.pyplot as plt

# Define a function to calculate positive polarity and group by quarter
def calculate_quarterly_polarity(df):
    # Convert the date to datetime if not already
    df['Date'] = pd.to_datetime(df['Date'])

    # Create a Quarter column
    df['Quarter'] = df['Date'].dt.to_period('Q')

    # Count sentiments grouped by Quarter
    sentiment_counts = df.groupby('Quarter')['Sentiment'].value_counts().unstack(fill_value=0)

    # Calculate Positive Polarity
    sentiment_counts['Positive Polarity'] = sentiment_counts.get('Positive', 0) / (
        sentiment_counts.get('Positive', 0) + sentiment_counts.get('Negative', 0)
    )

    # Reset index and return the result
    sentiment_counts.reset_index(inplace=True)
    return sentiment_counts[['Quarter', 'Positive Polarity']]

# Plot quarterly time-series data for each DataFrame
for i, df in enumerate(dataframes):
    try:
        # Calculate quarterly positive polarity
        quarterly_data = calculate_quarterly_polarity(df)

        # Plot the data
        plt.figure(figsize=(10, 6))
        plt.plot(quarterly_data['Quarter'].astype(str), quarterly_data['Positive Polarity'], marker='o')
        plt.title(f'Quarterly Positive Sentiment Polarity - DataFrame {i+1}')
        plt.xlabel('Quarter')
        plt.ylabel('Positive Polarity')
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        print(f"Plot created for DataFrame {i+1}")
    except Exception as e:
        print(f"Error processing DataFrame {i+1}: {e}")
```

```python
import pandas as pd

# Create a DataFrame to store all the positive polarity data
all_positive_polarity = pd.DataFrame(columns=["DataFrame", "Quarter", "Positive Polarity"])

# Loop through all DataFrames and calculate positive polarity
for i, df in enumerate(dataframes):
    try:
        # Ensure Date is in datetime format
        df['Date'] = pd.to_datetime(df['Date'])
        df['Quarter'] = df['Date'].dt.to_period('Q')

        # Group by Quarter and calculate sentiment counts
        sentiment_counts = df.groupby('Quarter')['Sentiment'].value_counts().unstack(fill_value=0)

        # Calculate positive polarity
        sentiment_counts['Positive Polarity'] = sentiment_counts.get('Positive', 0) / (
            sentiment_counts.get('Positive', 0) + sentiment_counts.get('Negative', 0)
        )

        # Reset index to prepare for table
        sentiment_counts.reset_index(inplace=True)

        # Add DataFrame number
        sentiment_counts["DataFrame"] = f"DataFrame {i+1}"

        # Append to the final table
        if not sentiment_counts.empty:
            all_positive_polarity = pd.concat(
                [all_positive_polarity, sentiment_counts[["DataFrame", "Quarter", "Positive Polarity"]]],
                ignore_index=True,
            )

    except Exception as e:
        print(f"Error processing DataFrame {i+1}: {e}")

# Reset index for the final table
all_positive_polarity.reset_index(drop=True, inplace=True)

# Display the table
print("Positive Polarity Table:")
print(all_positive_polarity)

# Save the table to a CSV file
output_file = "/content/positive_polarity_table.csv"
all_positive_polarity.to_csv(output_file, index=False)
print(f"Table saved to {output_file}")

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Create the dataset
data = {
    "Positive Polarity": [
        0.081206497, 0.081081081, 0.094339623, 0.122754491, 0.147147147, 0.115151515,
        0.092592593, 0.094339623, 0.093484419, 0.114503817, 0.103960396, 0.081424936,
        0.207612457, 0.083783784, 0.078817734, 0.132791328, 0.095057034, 0.171511628,
        0.129436326, 0.096385542
    ],
    "Net Change": [
        -205, 124, 76, 142, 154, 41, -390, 67, -93, 653, -170, -44, 292, -205,
        -44, -90, 155, 11, -92, 110
    ]
}
df = pd.DataFrame(data)

# Separate input (X) and output (y)
X = df["Positive Polarity"].values.reshape(-1, 1)
y = df["Net Change"].values.reshape(-1, 1)

# Normalize the data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Build the neural network model
model = Sequential([
    Dense(10, input_dim=1, activation='relu'),  # Input layer with 10 neurons
    Dense(10, activation='relu'),              # Hidden layer with 10 neurons
    Dense(1, activation='linear')              # Output layer with 1 neuron
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])

# Train the model
history = model.fit(X_train, y_train, epochs=200, batch_size=4, validation_data=(X_test, y_test), verbose=1)

# Evaluate the model
loss, mae = model.evaluate(X_test, y_test, verbose=0)
print(f"Mean Absolute Error on Test Data: {mae:.4f}")

# Predict on test data
y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)

# Rescale the test data back to original scale
y_test_rescaled = scaler_y.inverse_transform(y_test)

# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(scaler_X.inverse_transform(X_test), y_test_rescaled, color='blue', label='Actual Net Change')
plt.scatter(scaler_X.inverse_transform(X_test), y_pred, color='red', label='Predicted Net Change')
plt.title("Neural Network: Actual vs Predicted Net Change", fontsize=16)
plt.xlabel("Positive Polarity", fontsize=14)
plt.ylabel("Net Change", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()

# Plot training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.title("Training and Validation Loss", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
